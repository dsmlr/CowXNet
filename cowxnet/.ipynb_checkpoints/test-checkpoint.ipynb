```python
import os
import pandas as pd
import numpy as np
import core as CowXNet
```

```python
df = pd.read_pickle(os.path.join('..', '..', 'data', 'annotations.pkl'))
```

```python
for img in df['id']['image'][9:10]:

    img_path = os.path.join('..', '..', 'data', 'images', 'original', img + '.png')

    img_df = df[df['id']['image'] == img]

    boxes = []

    for i in range(len(img_df)):

        boxes.append({
            'box': (img_df.iloc[i]['cow-box']['xmin'], img_df.iloc[i]['cow-box']['ymin'], img_df.iloc[i]['cow-box']['xmax'], img_df.iloc[i]['cow-box']['ymax']),
            'keypoints': [
                (img_df.iloc[i]['head']['x'], img_df.iloc[i]['head']['y']), 
                (img_df.iloc[i]['body']['x'], img_df.iloc[i]['body']['y']), 
                (img_df.iloc[i]['tail-head']['x'], img_df.iloc[i]['tail-head']['y'])],
            'heat_label': img_df.iloc[i]['label']['heat_label'],
            'cow': img_df.iloc[i]['id']['cow']
        })

    CowXNet.visualize.visualize_cowxnet(img_path, boxes)
    
```
